```python
import pandas as pd
import numpy as np
```

```python
refugees = pd.read_csv('refugees.csv')
```

```python
refugees
```

```python
mapdata = pd.read_json('custom.json')
```

```python
mapnames = []
for country in mapdata.features:
    try:
        mapnames.append(str(country['properties']['name']))
    except:
        print country['properties']['name']

mapnames = pd.DataFrame(mapnames, columns=['name'])
mapnames.head()
```

```python
asylum = refugees['Country/territory of asylum/residence'].values
```

```python
origin = refugees['Origin'].values
```

```python
unhrc_countries = pd.Series(np.append(asylum,origin))
unhrc_countries = pd.DataFrame(unhrc_countries.unique(), columns=['unhrc'])
unhrc_countries.head()
```

```python
joined = unhrc_countries.merge(mapnames, how='outer', left_on='unhrc',right_on='name' )
```

```python
missing = joined[joined.isnull().any(axis=1)]
```

```python
missing.to_csv('missing.csv', index=False)
```
